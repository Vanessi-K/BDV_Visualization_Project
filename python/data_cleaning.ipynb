```python
import pandas as pd
import numpy as np
```

```python
# Loading raw data
data = pd.read_csv('../data/Crime_Data_from_2020_to_Present.csv')
data.head()
```

```python
data.describe()
```

```python
# Use numbers for the representation of the month in 'occ_month' column
data['occ_month'] = pd.to_datetime(data['DATE OCC']).dt.month

# Drop the first seven characters from TIME OCC
data['TIME OCC'] = data['TIME OCC'].astype(str).str[10:16]
```

```python
# Drop obsolete columns
columns_to_drop = ['LAT', 'LON', 'Crm Cd 1', 'Mocodes', 'LOCATION', 'Date Rptd', 'DATE OCC', 'Rpt Dist No']
data = data.drop(columns=columns_to_drop)

data.head()
```

```python
# Create a separate dataframe that contains the mapping of 'Crm Cd' to 'Crm Cd Desc'
crime_code_mapping = data[['Crm Cd', 'Crm Cd Desc']].drop_duplicates().reset_index(drop=True)
crime_code_mapping.head()
```

```python
# Create a separate dataframe that contains the mapping of 'AREA' to 'AREA NAME'
area_mapping = data[['AREA', 'AREA NAME']].drop_duplicates().reset_index(drop=True)
area_mapping.head()
```

```python
# Create a separate dataframe that contains the mapping of 'Premis Cd' to 'Premis Desc'
premis_code_mapping = data[['Premis Cd', 'Premis Desc']].drop_duplicates().reset_index(drop=True)
premis_code_mapping.head()
```

```python
# Create a separate dataframe that contains the mapping of 'Weapon Used Cd' to 'Weapon Desc'
weapon_code_mapping = data[['Weapon Used Cd', 'Weapon Desc']].drop_duplicates().reset_index(drop=True)
weapon_code_mapping.head()
```

```python
# Create a separate dataframe that contains the mapping of 'Status' to 'Status Desc'
status_mapping = data[['Status', 'Status Desc']].drop_duplicates().reset_index(drop=True)
status_mapping.head()
```

```python
# Drop description columns from the main data
columns_to_drop_desc = ['Crm Cd Desc', 'AREA NAME', 'Premis Desc', 'Weapon Desc', 'Status Desc']
data = data.drop(columns=columns_to_drop_desc)

data.head()
```

```python
# What are the entries in the 'Vict Sex' column?
data['Vict Sex'].unique()
```

```python
# Truncate 'Vict Sex' entries to the first character
data['Vict Sex'] = data['Vict Sex'].astype(str).str[0]
```

```python
# Save mapping tables to CSV files
crime_code_mapping.to_csv('../data/crime_code_mapping.csv', index=False)
area_mapping.to_csv('../data/area_mapping.csv', index=False)
premis_code_mapping.to_csv('../data/premis_code_mapping.csv', index=False)
weapon_code_mapping.to_csv('../data/weapon_code_mapping.csv', index=False)
status_mapping.to_csv('../data/status_mapping.csv', index=False)
```
