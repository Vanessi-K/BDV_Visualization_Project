```python
import pandas as pd
import numpy as np
```

```python
# Loading raw data
data = pd.read_csv('../data/Crime_Data_from_2020_to_Present.csv')
data.head()
```

```python
data.describe()
```

```python
# How many different types are there in the 'Crm Cd' column?
data['Crm Cd'].nunique()
```

```python
# Store cleaned data into a new variable
crm_data = data.copy()

# Only keep certain columns
columns_to_keep = ['Crm Cd', 'Crm Cd Desc', 'Part 1-2']
crm_data = crm_data[columns_to_keep]

# Count all rows that are the same and store the counts in a new column 'Count'
crm_data['Count'] = 1
crm_data = crm_data.groupby(['Crm Cd', 'Crm Cd Desc', 'Part 1-2']).count().reset_index()

# Order by 'Count' column in ascending order
crm_data = crm_data.sort_values(by='Count', ascending=True).reset_index(drop=True)

crm_data
```
